```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Input

model = Sequential()
model.add(Input(shape=(4,))) # input layer : 각 샘플이 길이 4인 벡터(특징 4개)
model.add(Dense(50, activation="sigmoid")) # hidden layer
model.add(Dense(30, activation="sigmoid")) # hidden layer
model.add(Dense(3, activation="softmax")) # output layer
# model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
path = 'model-{epoch:02d}-{val_accuracy:.4f}.keras'
checkpoint = ModelCheckpoint(path, save_best_only=True, monitor='val_accuracy', verbose=1)
```

```python
from tensorflow.keras.callbacks import TensorBoard
tensorboard = TensorBoard(log_dir='logs', histogram_freq=1, embeddings_freq=1)
```

```python
model.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=50, callbacks=[tensorboard])
```
